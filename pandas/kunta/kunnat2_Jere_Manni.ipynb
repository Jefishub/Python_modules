# Pandas Excercise


In this excercise we modify a dataset and answer few questions
- What is the total population in capital region (Helsinki, Espoo, Vantaa, Kauniainen)
- What is the muncipality where the minimum population resides
- Where is the highest Työllisyysaste, %
- Where is the lowest Työllisyysaste, %

I am using pivot table to handle data

Tässä tehtävässä muokataan ensin excel-talukko sopivaan muotoon. 
Taulukon tilastot pitää muuttaa ensin pivot-taulukoksi, 
jotta dataa voidaan sen jälkeen analysoida.

```python
import pandas as pd
```

# DataFramen muokkaus

```python
df = pd.read_excel("Kunnat2.xlsx", header=2, skipfooter=46)
```

```python
df
```

```python
df.fillna(method="ffill", inplace=True)
df
```

```python
df.rename(columns={"Unnamed: 0":"Kaupunki","Unnamed: 1":"Tilasto"}, inplace=True)
df
```

# Kaupunkien ja kuntien erottelu

Erotellaan kaupungit, seutukunnat sekä maakunnat. Käytetään pivot taulukointia tilastojen analysoimiseen

```python
#erotellaan kaupungit
kaupungit = df.head(1244)
kaupungit = kaupungit.pivot(index="Kaupunki", columns="Tilasto", values="2017")
kaupungit
```

```python
#erotellaan seutukunnat
seutukunnat = df.iloc[1320:]
seutukunnat = seutukunnat.pivot(index="Kaupunki", columns="Tilasto", values="2017")
seutukunnat
```

```python
#erotellaan maakunnat
maakunnat = df.iloc[1244:1320]
maakunnat = maakunnat.pivot(index="Kaupunki", columns="Tilasto", values="2017")
maakunnat
```

# Tilastojen analysointi

```python
#Population of capital region
kaupungit.loc[["Helsinki","Espoo", "Vantaa", "Kauniainen"]]["Väkiluku"]
```

```python
#Population of capital region
print("Capital region population:",kaupungit.loc[["Helsinki","Espoo", "Vantaa", "Kauniainen"]]["Väkiluku"].sum())
```

```python
#What is the municipality where the minimun population resides
maakunnat.sort_values(by=["Väkiluku"],ascending=False)["Väkiluku"]
```

```python
maakunnat.sort_values(by=["Työllisyysaste, %"],ascending=False)["Työllisyysaste, %"]
```

```python
kaupungit.sort_values(by=["Työllisyysaste, %"],ascending=False)["Työllisyysaste, %"]
```
